```python
import importlib.util
import sys
from pathlib import Path

# Path to db.py
db_path = Path(r"C:\Users\Hp\nidah_portal\backend\app\db.py")

spec = importlib.util.spec_from_file_location("db", db_path)
db = importlib.util.module_from_spec(spec)
sys.modules["db"] = db
spec.loader.exec_module(db)

# Now you can use Base from db
Base = db.Base

print("db module loaded successfully:", db)
```
